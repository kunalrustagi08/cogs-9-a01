```python
# Importing libraries like pandas, numpy, sckit learn
import numpy as np
import random
import pandas as pd
from sklearn.utils import shuffle
```

```python
# Reading csv file with student emails, names and section id
df = pd.read_csv("/content/student_pids - Sheet1.csv", usecols=['Sec ID', 'Student', 'Email'])
df
```

```python
# Check if unique
df['Email'].is_unique
```

```python
# Define section id for A01
section_id = 99956
df[df['Sec ID'] == section_id]
```

```python
# Selecting only A01 students
new_df = df[df['Sec ID'] == section_id]

# Random shuffling of students
df_shuffled = shuffle(new_df)
df_shuffled = df_shuffled.reset_index(drop=True)
```

```python
# Team names generated by Chat GPT-3 based on cognitive science. Feel free to edit your team names
team_names = ['Mind Melders',
                'Brain Busters',
                'Cognitive Crusaders',
                'Perception Pioneers',
                'Neural Navigators',
                'Memory Mavericks',
                'Consciousness Creators',
                'Thinking Titans',
                'Behavioral Bandits',
                'Emotion Engineers',
                'Decision Dynamics',
                'Attention Avengers',
                'Language Luminaries',
                'Neuro-Networkers',
                'Mental Marvels',
            ]

grouped = df_shuffled.groupby(df_shuffled.index // 5)
```

```python
# Grouping into 5 and assigning team numbers
df_shuffled['team'] = grouped.grouper.group_info[0]
df_shuffled = df_shuffled.sort_values(by = 'team')
df_shuffled = df_shuffled.reset_index(drop=True)
df_shuffled
```

```python
# Assigning team names
df_shuffled['team'] = df_shuffled['team'].apply(lambda x: team_names[x])
df_shuffled
```

```python
# Rearranging columns
cols = df_shuffled.columns.tolist()
cols = cols[-1:] + cols[1:-1]
df = df_shuffled[cols]
df
```

```python
# Extracting final data in new csv
df.to_csv('a01.csv')
```
